# Monte Carlo Estimation of $\pi$

## Introduction

The constant $\pi$ is one of the most fundamental numbers in mathematics.
It appears in geometry, physics, probability theory and many computational algorithms.

One interesting way to estimate the value of $\pi$ is the **Monte Carlo method**.
This method uses random sampling to approximate numerical values.

In this project we implement a Monte Carlo simulation in Python to estimate the value of $\pi$ and analyze how the accuracy improves as the number of random samples increases.

## Mathematical Background

The Monte Carlo method estimates the value of $\pi$ using random sampling.

Consider a square with side length $2$ and a circle with radius $1$ inside it.

The area of the circle is

$$
A = \pi r^2
$$

Since $r = 1$:

$$
A = \pi
$$

The area of the square is

$$
A = (2)^2 = 4
$$

If we randomly generate points inside the square, some of them will fall inside the circle.

The ratio between points inside the circle and total points approximates the ratio between the circle area and the square area:

$$
\frac{\text{points inside circle}}{\text{total points}} \approx \frac{\pi}{4}
$$

From this we obtain the Monte Carlo approximation:

$$
\pi \approx 4 \times \frac{\text{points inside circle}}{\text{total points}}
$$

## Monte Carlo Estimator

Let

- $N$ be the total number of generated points
- $N_c$ be the number of points inside the circle

The Monte Carlo estimator for $\pi$ is:

$$
\hat{\pi} = 4 \frac{N_c}{N}
$$

As $N$ increases, the estimator $\hat{\pi}$ converges to the true value of $\pi$.

## Python Implementation

We now implement the Monte Carlo algorithm in Python. The algorithm generates random points inside the square and counts how many fall inside the unit circle.

The estimate of $\pi$ is then computed using the Monte Carlo estimator.

```python
def estimate_pi(n_points):

    inside_circle = 0

    for _ in range(n_points):

        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        if x**2 + y**2 <= 1:
            inside_circle += 1

    return 4 * inside_circle / n_points
```

## Experiments

To evaluate the accuracy of the Monte Carlo method we run simulations with different numbers of random points.

As the number of generated points increases, the estimate of $\pi$ should become more accurate.

```python
estimate_pi(100000)
```

```python
points = [100, 1000, 10000, 100000]

for n in points:
    print("Points:", n, "Estimated π:", estimate_pi(n))
```

## Visualization of the Simulation

To better understand how the Monte Carlo method works, we visualize the randomly generated points.

Points inside the circle represent successful samples, while points outside the circle fall outside the unit circle.

```python
def monte_carlo_visualization(n_points):

    x_inside = []
    y_inside = []

    x_outside = []
    y_outside = []

    for _ in range(n_points):

        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        if x**2 + y**2 <= 1:
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)

    plt.figure(figsize=(6,6))

    plt.scatter(x_inside, y_inside, s=1, label="Inside Circle")
    plt.scatter(x_outside, y_outside, s=1, label="Outside Circle")

    plt.title(r"Monte Carlo Estimation of $\pi$")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.axis("equal")
    plt.show()
```

```python
monte_carlo_visualization(5000)
```

## Conclusion

The Monte Carlo method demonstrates how randomness can be used to approximate mathematical constants.

As the number of sampled points increases, the estimate of $\pi$ becomes closer to its true value:

$$
\pi \approx 3.141592653589793
$$

Although the Monte Carlo method is not the most efficient way to compute $\pi$, it is an excellent example of probabilistic simulation and numerical approximation.
